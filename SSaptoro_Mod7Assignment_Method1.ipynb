```python
import numpy as np
np.set_printoptions(precision=15,suppress=True)
```

```python
# First I/O pair
first_i1=1
first_i2=1
first_d=0.9

# Second I/O pair
second_i1=-1
second_i2=-1
second_d=0.05

# initialize weights from inputs 1 and 2 to nodes 1 and 2
w_i1_1=0.3
w_i2_1=0.3
w_i1_2=0.3
w_i2_2=0.3
bias1=0
bias2=0

#  initialize weights from nodes 1 and 2 to node 3
w_1_3=.8
w_2_3=.8
bias3=0

eta=1

error = 1
counter = 0
```

```python
def activation(A):
    return 1/(1+np.exp(-1*A))
```

```python
# while abs(error) > 0.0001:
for i in range(0,15):
    print("Iteration", counter+1)
    
    activity_1 = w_i1_1*first_i1 + w_i2_1*first_i2 + bias1
    y_1 = activation(activity_1)
    #print("First I/O activation value for node 1 is", y_1)
    activity_2 = w_i1_2*first_i1 + w_i2_2*first_i2 + bias2
    y_2 = activation(activity_2)
    #print("First I/O activation value for node 2 is", y_2)
    activity_3 = w_1_3*y_1 + w_2_3*y_2 + bias3
    y_3 = activation(activity_3)
    print("Output for first I/O is", y_3)
    
    error = first_d-y_3
    #big_error = .5*error**2
    #print("First I/O Error is", big_error)
    
    delta_3 = y_3*(1-y_3)*error
    w_1_3_former = w_1_3
    w_2_3_former = w_2_3
    w_1_3 = w_1_3 + eta*delta_3*y_1
    w_2_3 = w_2_3 + eta*delta_3*y_2
    bias3 = bias3 + eta*delta_3
    print("New weights into node 3 after first I/O are:", w_1_3, "(from node 1) and", w_2_3, "(from node 2)")
    
    delta_2 = y_2*(1-y_2)*delta_3*w_2_3_former
    w_i1_2 = w_i1_2 + eta*delta_2*first_i1
    w_i2_2 = w_i2_2 + eta*delta_2*first_i2
    bias2 = bias2 + eta*delta_2
    print("New weights into node 2 after first I/O are:", w_i1_2, "(from input 1) and", w_i2_2, "(from input 2)")
    
    delta_1 = y_1*(1-y_1)*delta_3*w_1_3_former
    w_i1_1 = w_i1_1 + eta*delta_1*first_i1
    w_i2_1 = w_i2_1 + eta*delta_1*first_i2
    bias2 = bias2 + eta*delta_2
    print("New weights into node 1 after first I/O are:", w_i1_1, "(from input 1) and", w_i2_1, "(from input 2)")
    
    print("\n")
        
    activity_1 = w_i1_1*second_i1 + w_i2_1*second_i2 + bias1
    y_1 = activation(activity_1)
    #print("Second I/O activation value for node 1 is", y_1)
    activity_2 = w_i1_2*second_i1 + w_i2_2*second_i2 + bias2
    y_2 = activation(activity_2)
    #print("Second I/O activation value for node 2 is", y_2)
    activity_3 = w_1_3*y_1 + w_2_3*y_2 + bias3
    y_3 = activation(activity_3)
    print("Output for second I/O is", y_3)
    
    error = second_d-y_3
    #big_error = .5*error**2
    #print("Second I/O Error is", big_error)
    
    delta_3 = y_3*(1-y_3)*error
    w_1_3_former = w_1_3
    w_2_3_former = w_2_3
    w_1_3 = w_1_3 + eta*delta_3*y_1
    w_2_3 = w_2_3 + eta*delta_3*y_2
    bias3 = bias3 + eta*delta_3
    print("New weights into node 3 after second I/O are:", w_1_3, "(from node 1) and", w_2_3, "(from node 2)")
    
    delta_2 = y_2*(1-y_2)*delta_3*w_2_3_former
    w_i1_2 = w_i1_2 + eta*delta_2*second_i1
    w_i2_2 = w_i2_2 + eta*delta_2*second_i2
    bias2 = bias2 + eta*delta_2
    print("New weights into node 2 after second I/O are:", w_i1_2, "(from input 1) and", w_i2_2, "(from input 2)")
    
    delta_1 = y_1*(1-y_1)*delta_3*w_1_3_former
    w_i1_1 = w_i1_1 + eta*delta_1*second_i1
    w_i2_1 = w_i2_1 + eta*delta_1*second_i2
    bias2 = bias2 + eta*delta_2
    print("New weights into node 1 after second I/O are:", w_i1_1, "(from input 1) and", w_i2_1, "(from input 2)")
    
    counter = counter + 1
    print("\n")
    print("\n")
```

```python
activity_1 = w_i1_1*first_i1 + w_i2_1*first_i2 + bias1
y_1 = activation(activity_1)
#print("First I/O activation value for node 1 is", y_1)
activity_2 = w_i1_2*first_i1 + w_i2_2*first_i2 + bias2
y_2 = activation(activity_2)
#print("First I/O activation value for node 2 is", y_2)
activity_3 = w_1_3*y_1 + w_2_3*y_2 + bias3
y_3 = activation(activity_3)
print("Output for first I/O is", y_3)
first_error = first_d-y_3
big_error = .5*first_error**2
print("First I/O Error is", big_error)

activity_1 = w_i1_1*second_i1 + w_i2_1*second_i2 + bias1
y_1 = activation(activity_1)
#print("First I/O activation value for node 1 is", y_1)
activity_2 = w_i1_2*second_i1 + w_i2_2*second_i2 + bias2
y_2 = activation(activity_2)
#print("First I/O activation value for node 2 is", y_2)
activity_3 = w_1_3*y_1 + w_2_3*y_2 + bias3
y_3 = activation(activity_3)
print("Output for second I/O is", y_3)
second_error = second_d-y_3
big_error = .5*second_error**2
print("Second I/O Error is", big_error)
```
